```python
import re
def clean_arabic_text(text):
    text = re.sub(r'Page number: \d+', '', text)
    text = re.sub(r'- \d+ -', '', text)
    text = re.sub(r'\b\d+\b', '', text)
    text = re.sub(r'[^\u0600-\u06FF\s]', '', text)
    text = re.sub(r'[^\w\s\u0600-\u06FF]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text





final['context'] = final['context'].apply(clean_arabic_text)
final['question'] = final['question'].apply(clean_arabic_text)
final['answer'] = final['answer'].apply(clean_arabic_text)



```

```python


print("Context:\n", final.loc[2000, "context"])

print("Question:\n", final.loc[2000, "question"])
print("Answer:\n", final.loc[2000, "answer"])
```
